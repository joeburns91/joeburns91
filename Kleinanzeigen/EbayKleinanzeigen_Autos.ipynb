Looking above we can see that we have some data entries in registration_year (the first year the car was registered) that are invalid. We have dates in the future, and dates that predate the autommobile. We can also see our average year of first registration is 2005, including these false data points. We will only select vehicles listed from the 1950s to 2016. This is because there are almost zero cars older than 1950 on the road, and our latest date of scraping was april 07 2016.

```python
registration_year_filtered = autos_filtered[autos_filtered["registration_year"].between(int(1950),int(2016))]
registration_year_filtered.describe()
```

Here we can see that we were able to bring our standard deviation for vehicle registration dates down from 89 to only 7, all while only removing 2000 etries (4%). All of these removed were either impossible, or highly highly improbably to be correct.

```python
autos_filtered = registration_year_filtered
print(autos_filtered['brand'].unique())
auto_brands = autos_filtered['brand']
print(auto_brands.value_counts(normalize = True))
```

Here we will select VW, bmw, opel, benz, audi, and ford as they are all over 5% of the total listings remaining.

```python
price_dict = {}
agg_brands = ['volkswagen', 'bmw', 'opel', 'mercedes_benz', 'audi', 'ford']
for element in agg_brands:
    price_mean = autos_filtered.loc[autos_filtered["brand"] == element, "price"].mean()
    price_mean = round(price_mean, 2)
    price_dict[element] = price_mean
print(price_dict)
```

Here we can see that there are roughly 2 classes of automobiles. Benz, bmw, and audi are a price step above ford, opel, and VW.

```python
mileage_dict = {}
for element in agg_brands:
    mileage_mean = autos_filtered.loc[autos_filtered["brand"] == element, "odometer_km"].mean()
    mileage_mean = round(mileage_mean * 0.621371)
    mileage_dict[element] = mileage_mean
print(mileage_dict)
```

Here we see that the average mileages are roughly equivalent, giving more support to the price comparison in the cell made above.

```python
bmp_series = pd.Series(price_dict, name = 'mean_price')
bmm_series = pd.Series(mileage_dict, name = 'mean_miles')

overall_mean_price = bmp_series.mean()
overall_mean_mileage = bmm_series.mean()

print('Mean price of all vehicles: ' + str(round(overall_mean_price, 2)))
print('Mean mileage of all vehicles: ' + str(round(overall_mean_mileage)))
print(bmp_series)
#print(bmm_series)
```
